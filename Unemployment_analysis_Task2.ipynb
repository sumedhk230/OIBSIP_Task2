```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('/content/Unemployment in India.csv')
df = pd.read_csv('/content/Unemployment_Rate_upto_11_2020.csv')
df.head()
```

```python
df.info()
```

Renaming the column names by using .columns method 

```python
df.columns = ['region','date','freq','unrate','employed','labour participation rate','direction','longitude','latitude']
```

```python
df.info()
```

```python
df.describe()
```

#check if there are any null or missing values in this dataset

```python
df.isnull().sum()
```

A simple 2d plot using seaborn for estimating unemployment rate in india with respect to end dates of all the months from 2020 year

```python
plt.figure(figsize = (12,10))
sns.lineplot(data = df,x = 'date',y = 'unrate',color = 'green',ci =None)
```

A simple 2d plot using seaborn for estimating unemployment rate in india with respect to all the regions(states) in india

```python
plt.figure(figsize=(50,20))
sns.lineplot(data = df,x = 'region',y = 'unrate',color = 'red',ci =None)
```

Here, We can say that the number of people employed in Gujarat is much higher than that of Haryana and Himachal Pradesh.  Himachal Pradesh due to its less population has very less people employed with respect to Gujarat and Haryana

```python
plt.figure(figsize = (50,20))
sns.lineplot(data = df[(df.region >= 'Gujarat')&(df.region <= 'Himachal Pradesh')], x = 'region', y = 'employed',ci =None)
```

```python
dataplot = sns.heatmap(df.corr(),cmap = "PiYG")
plt.show()
```

```python
sns.pairplot(df)
plt.show
```

```python
plt.title("Number of Employed per region")
sns.histplot(x="employed", hue="direction", data=df)
plt.show()
```

```python
plt.title("Unemployment rate per region")
sns.histplot(x="unrate", hue="direction", data=df)
plt.show()
```

```python
un = df[["region", "direction", "unrate"]]
figure = px.sunburst(un, path=["direction", "region"], 
                     values="unrate", 
                     width=1000, height=1000, color_continuous_scale="RdY1Gn", 
                     title="A sunburst Plot for Unemployment Rates in India")
figure.show()
```
